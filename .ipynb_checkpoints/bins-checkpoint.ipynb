```python
def standardize_pay(num):
    if num > 975 and num <= 10424:
        output = 976.
    elif num > 10424 and num <= 19873:
        output = 10424.
    elif num > 19873 and num <= 29322:
        output = 19873.
    elif num > 29322 and num <= 38771:
        output = 29322.
    elif num > 38771 and num <= 48220:
        output = 38771.
    elif num > 48220 and num <= 57669:
        output = 48220.
    elif num > 57669 and num <= 67118:
        output = 57669.
    elif num > 67118 and num <= 76568:
        output = 67118.
    elif num > 76568 and num <= 86017:
        output = 76568.
    elif num > 86017 and num <= 95466:
        output = 86017.
    elif num > 95466 and num <= 104915:
        output = 95466.
    elif num > 104915 and num <= 114364:
        output = 104915.
    elif num > 114364 and num <= 123813:
        output = 114364.
    elif num > 123813 and num <= 133262:
        output = 123813.
    elif num > 133262 and num <= 142712:
        output = 133262.
    elif num > 142712 and num <= 152161:
        output = 142712.
    elif num > 152161 and num <= 161610:
        output = 152161.
    elif num > 161610 and num <= 171059:
        output = 161610.
    elif num > 171059 and num <= 180508:
        output = 171059.
    elif num > 180508 and num <= 189957:
        output = 180508.
    elif num > 189957 and num <= 199406:
        output = 189957.
    elif num > 199406 and num <= 208856:
        output = 199406.
    elif num > 208856 and num <= 218305:
        output = 208856.
    elif num > 218305 and num <= 227754:
        output = 218305.
    elif num > 227754 and num <= 237203:
        output = 22754.
    elif num > 237203 and num <= 246652:
        output = 237203.
    elif num > 246652 and num <= 256101:
        output = 246652.
    elif num > 256101 and num <= 265550:
        output = 256101.
    elif num > 265550 and num <= 275000:
        output = 265550.
    else:
        output = np.nan
    return output
```

```python
df['pay'] = df['Gross Pay'].apply(standardize_pay)
df['pay'].unique()
```
